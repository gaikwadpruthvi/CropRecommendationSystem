### Seperating features and target label

```python
features = df[['N', 'P','K','temperature', 'humidity', 'ph', 'rainfall']]
target = df['label']
#features = df[['temperature', 'humidity', 'ph', 'rainfall']]
labels = df['label']
```

```python
# Initialzing empty lists to append all model's name and corresponding name
acc = []
f1=[]
rec=[]
time=[]
pre=[]
model = []
```

```python
# Splitting into train and test data

from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(features,target,test_size = 0.2,random_state =2)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from timeit import default_timer as timer

start = timer()

print(23*2.3)

DecisionTree = DecisionTreeClassifier(criterion="entropy",random_state=2,max_depth=5)

DecisionTree.fit(Xtrain,Ytrain)

predicted_values = DecisionTree.predict(Xtest)
x = metrics.accuracy_score(Ytest, predicted_values)
f = metrics.f1_score(Ytest, predicted_values,average='micro')
r=metrics.recall_score(Ytest, predicted_values,average='micro')
p=metrics.precision_score(Ytest, predicted_values,average='micro')
acc.append(x)
f1.append(f)
rec.append(r)
pre.append(p)
model.append('Decision Tree')
print("DecisionTrees's Accuracy is: ", x*100)

print(classification_report(Ytest,predicted_values))


end = timer()

t=end - start
time.append(t)
```

```python
# Generate predictions with the best model

# Create the confusion matrix
cm = confusion_matrix(Ytest, predicted_values)

ConfusionMatrixDisplay(confusion_matrix=cm).plot(cmap=plt.cm.Blues)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# Cross validation score (Decision Tree)
score = cross_val_score(DecisionTree, features, target,cv=5)
```

```python
score
```

```python
import pickle
# Dump the trained Naive Bayes classifier with Pickle
DT_pkl_filename = 'DecisionTree.pkl'
# Open the file to save as pkl file
DT_Model_pkl = open(DT_pkl_filename, 'wb')
pickle.dump(DecisionTree, DT_Model_pkl)
# Close the pickle instances
DT_Model_pkl.close()
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler
# fit scaler on training data
from timeit import default_timer as timer

start = timer()

print(23*2.3)
norm = MinMaxScaler().fit(Xtrain)
X_train_norm = norm.transform(Xtrain)
# transform testing dataabs
X_test_norm = norm.transform(Xtest)
SVM = SVC(kernel='poly', degree=3, C=1)
SVM.fit(X_train_norm,Ytrain)
predicted_values = SVM.predict(X_test_norm)
x = metrics.accuracy_score(Ytest, predicted_values)
f = metrics.f1_score(Ytest, predicted_values,average='micro')
r=metrics.recall_score(Ytest, predicted_values,average='micro')
p=metrics.precision_score(Ytest, predicted_values,average='micro')
acc.append(x)
f1.append(f)
rec.append(r)
pre.append(p)
model.append('SVM')
print("SVM's Accuracy is: ", x)

print(classification_report(Ytest,predicted_values))


end = timer()

t=end - start
time.append(t)
```

```python
# Cross validation score (SVM)
score = cross_val_score(SVM,features,target,cv=5)
score
```

```python
cm = confusion_matrix(Ytest, predicted_values)

ConfusionMatrixDisplay(confusion_matrix=cm ).plot(cmap=plt.cm.Blues)
```
